### Gas Price

Price over time for the history of this car's fill-up

```python
prices = df['Gas Price']
date = df['Date']

%matplotlib inline
plt.figure(figsize=(20,10))
plt.plot(date,prices,'*-g')
plt.title("Gas Price over Time",fontsize=20)
plt.xlabel("Time",fontsize=14)
plt.xticks(rotation=90)
x_ticks = np.arange(0, len(df), 2)
plt.xticks(df['Date'][x_ticks])
plt.ylabel("Gas Price in Dollars",fontsize=14)
```

The price steadily increased until it reached around \\$3.75 and drops for a little bit then increases back up to around \$3.75 again with a steep drop afterwards

```python
mpg = df['Miles/Gallon']
date = df['Date']

%matplotlib inline
plt.figure(figsize=(20,10))
plt.plot(date,mpg,'-o')
plt.title("Miles Per Gallon",fontsize=20)
plt.xlabel("Time",fontsize=14)
plt.xticks(rotation=90)
x_ticks = np.arange(0, len(df), 2)
plt.xticks(df['Date'][x_ticks])
plt.ylabel("Miles Per Gallon",fontsize=14)
```

MPG floats around 30 to 35 mileage per gallon with the max being 39 and the min 19.

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = df_filtered[['Miles Before Fill-up']]
y = df_filtered[['Miles Left']]

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

print("Coefficent:",model.coef_)
print("Intercept:",model.intercept_)

rmse = np.sqrt(mean_squared_error(y,y_pred))
r2 = r2_score(y,y_pred)
print("Root Mean Square Erro:",rmse)
print("R-Squared:",r2)

plt.figure(figsize=(20,10))
plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r',label="y = -0.72x+338.19")
plt.title('Miles Driven vs Miles Left with Linear Regression',fontsize=20)

plt.text(425, 210, 'RMSE = %0.4f' % rmse)
plt.text(425, 200, 'R-squared = %0.4f' % r2)

plt.legend()
plt.xlabel("Miles Driven")
plt.ylabel("Miles Left")
```
