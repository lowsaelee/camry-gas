```python
df_filtered = df.dropna()
df_filtered
```

```python
df_filtered.shape
```

```python
milesDriven = df_filtered['Miles Before Fill-up']
milesLeft = df_filtered['Miles Left']

plt.scatter(milesDriven,milesLeft)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = df_filtered[['Miles Before Fill-up']]
y = df_filtered[['Miles Left']]

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

print("Coefficent:",model.coef_)
print("Intercept:",model.intercept_)

rmse = np.sqrt(mean_squared_error(y,y_pred))
r2 = r2_score(y,y_pred)
print("Root Mean Square Erro:",rmse)
print("R-Squared:",r2)

plt.figure(figsize=(20,10))
plt.scatter(x, y, s=10)
plt.plot(x, y_pred, color='r',label="y = -0.72x+338.19")
plt.title('Miles Driven vs Miles Left with Linear Regression',fontsize=20)

plt.text(425, 210, 'RMSE = %0.4f' % rmse)
plt.text(425, 200, 'R-squared = %0.4f' % r2)

plt.legend()
plt.xlabel("Miles Driven")
plt.ylabel("Miles Left")
```
